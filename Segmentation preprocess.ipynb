```python
import os
from PIL import Image
import cv2 as cv
```

```python
cwd = 'C:/Users/Asus/COVID-19 for segmentation/images/'
```

```python
for file in os.listdir(cwd):
    if(file[:-4] == 'jpeg'):
        Image.open(cwd+file).resize((224,224)).save(os.path.join(cwd, file[:-5] + '.png'))
    else:
        Image.open(cwd+file).resize((224,224)).save(os.path.join(cwd, file[:-4] + '.png'))
```

```python
cwd = 'C:/Users/Asus/COVID-19 for segmentation/masks/'
```

```python
for file in os.listdir(cwd):
    img = cv.imread(cwd+file)
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    out = Image.fromarray(img)
    out.save(cwd+file)
```
