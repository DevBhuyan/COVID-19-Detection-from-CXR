```python
from cv2 import createCLAHE
import cv2 as cv
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
clahe = createCLAHE(clipLimit = 5)
```

```python
cwd = 'C:/Users/Asus/COVID19 datasets (CLAHE clipLimit5)'
```

```python
cwd
```

```python
files = []
for dirname in os.listdir(cwd):
    cd = cwd+'/'+dirname
    for sub in os.listdir(cd):
        subd = cd+'/'+sub
        for subsub in os.listdir(subd):
            subsubd = subd+'/'+subsub
            for file in os.listdir(subsubd):
                fullfile = subsubd+'/'+file
                if(file[:5] == 'CLAHE'):
                    continue
                else:
                    try:
                        img = cv.imread(fullfile)
                        img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
                        clahe = createCLAHE(clipLimit = 5)
                        out = clahe.apply(img)
                        out = Image.fromarray(out)
                        out.save(subsubd+'/CLAHE'+file)
                        print(subsubd+'/CLAHE'+file)
                    except:
                        pass
```
